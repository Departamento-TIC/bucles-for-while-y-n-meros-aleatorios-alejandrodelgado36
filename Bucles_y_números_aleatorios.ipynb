#6. Escribe un programa en Python que:
* Le pida al usuario siete números
* Imprima la suma total de esos números
* Imprima la cuenta de las entradas positivas, las que sean iguales a cero, y las negativas. Emplea una cadena if, elif, else, en lugar de tres if consecutivos

````python
print('Dame siete números:')
np = 0
nn = 0
n0 = 0
nt = 0
for i in range(7):
  n = int(input('Dime un número: '))
  if n > 0:
    np += 1
  elif n == 0:
    n0 += 1
  else:
    nn += 1
  nt += n
print('La suma total es:', nt)
print('Hay', np,'entradas positivas.')
print('Hay', nn,'entradas negativas.')
print('Hay', n0,'entradas que igualen a 0.')
````

#8. Escribe un programa para jugar a piedra, tijera y papel:

* Crea un programa que imprima aleatoriamente 0, 1, ó 2.
* Usando sentencias if, expande el programa de manera que ahora imprima al azar piedra, papel o tijera. No hagas la selección desde una lista, tal como hemos visto en el capítulo.
* Añade al programa la opción de que primero le pregunte al usuario qué es lo que elige.(Sería más fácil darle a escoger entre las opciones 1, 2, ó 3.)
* Añade una declaración condicional para determinar quién gana.


````python
import random
print('Piedra, papel o tijera')
na = random.randrange(3)
if na == 0:
  resultado = 'Piedra'
elif na == 1:
  resultado = 'Papel'
else:
  resultado = 'Tijera'
print('1. Piedra')
print('2. Papel')
print('3. Tijera')
oj = int(input('Dime cual escojes: '))
if oj == 1:
  oj = 'Piedra'
elif oj == 2:
  oj = 'Papel'
elif oj == 3:
  oj = 'Tijera'
else:
  print('Opción no disponible.')
if oj == 'Piedra' and resultado == 'Tijera':
  print('Piedra le gana a tijera, ganaste!')
elif oj == resultado:
  print('Ambos escogieron ', oj ,'.Empate')
elif oj == 'Piedra' and resultado == 'Papel':
  print('Papel le gana a piedra, perdiste.')
elif oj == 'Papel' and resultado == 'Piedra':
  print('Papel le gana a piedra, ganaste!')
elif oj == 'Papel' and resultado == 'Tijera':
  print('Tijera le gana a papel, perdiste.')
elif oj == 'Tijera' and resultado == 'Papel':
  print('Tijera le gana a papel, ganaste!')
elif oj == 'Tijera' and resultado == 'Piedra':
  print('Piedra le gana a tijera, perdiste.')
else:
  print('Algo salío mal, inténtelo otra vez.')
````
